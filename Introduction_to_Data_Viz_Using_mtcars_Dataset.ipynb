It should also be clear by how `plotnine` treats these variables.

1. Create side-by-side boxplots of the `mpg` variable by the different values of the `am` variable. What happens if you convert the `am` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

When am is in string format, it is treated as a categorical variable, and each unique value of am forms a distinct category on the x-axis. In this case, the boxplots clearly represent the distribution of mpg values for each category of am.

When am is converted back to float format, it is treated as a numeric variable. The key difference is in how the variable am is interpreted: as categorical when it is a string and as numeric when it is a float, but this difference doesn’t have a significant impact on the visual representation. 


2. Create overlaid histograms of the `hp` variable for the different values of the `cyl` variable. What happens if you convert the `cyl` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

When cyl is converted back to a float and plotted as a factor (factor(cyl)), it is still treated as a categorical variable, and each unique numeric value of cyl forms a distinct category represented by different colors in the overlaid histograms. So visually, there is not much difference between the two plots in this specific case, as cyl inherently represents distinct categories.

```python
# Create plot for (1) here
#plot number 1 for pt 1
from plotnine import ggplot, aes, geom_boxplot, labs

# Create side-by-side boxplots with 'am' as string using plotnine
df['am'] = df['am'].astype(str)
plot_am_string = (
    ggplot(df, aes(x='am', y='mpg',fill="am")) +
    geom_boxplot() +
    labs(title='Boxplot of mpg by am (am as string)',
         x='am', y='mpg')
)

print(plot_am_string)

#plot number 2 for pt 1

# Create side-by-side boxplots with 'am' as float using plotnine
# Use 'factor(am)' to treat 'am' as a categorical variable
plot_am_float = (
    ggplot(df, aes(x='factor(am)', y='mpg',fill="am")) +  
    geom_boxplot() +
    labs(title='Boxplot of mpg by am (am as float)',
         x='am', y='mpg')
)
print(plot_am_float)
```

```python
# Create plot for (2) here
#plot number 2 for pt 1
from plotnine import ggplot, aes, geom_histogram, labs
# Create overlaid histograms with 'cyl' as string using plotnine
df['cyl'] = df['cyl'].astype(str)
plot_cyl_string = (
    ggplot(df, aes(x='hp', fill='cyl')) +
    geom_histogram(position='identity', alpha=0.6, bins=15, color='k') +
    labs(title='Overlaid Histograms of hp by cyl (cyl as string)',
         x='hp', y='Frequency')

)

print(plot_cyl_string)

#plot number 2 for pt 2

# Use 'factor(cyl)' to treat 'cyl' as a categorical variable
df['cyl'] = df['cyl'].astype(float)
plot_cyl_float = (
    ggplot(df, aes(x='hp', fill='factor(cyl)')) +
    geom_histogram(position='identity', alpha=0.6, bins=15, color='k') +
    labs(title='Overlaid Histograms of hp by cyl (cyl as float)',
         x='hp', y='Frequency')
)

print(plot_cyl_float)
```

Some of these variables, like the `cyl` variable, have numeric values that actually make sense as numbers (i.e. the number of cylinders in the engine). However, it doesn't make the most sense to "do math" with this type of variable (e.g. take averages and such) because there are so few different value this can take on AND they're an explicit choice made by the car manufacturer. So, it makes more sense to treat the `cyl` variable as a categorical variable despite it having numeric values.

You will need to keep these kinds of nuances about data in mind as you work with an increasing variety and richness of data, and do more complex things with them.

Choose 3 new `geometries` from the [data-to-viz website](https://www.data-to-viz.com/) for graphs that you'd like to explore using the `mtcars` dataset, and then create those graphs!

```python
# Create plot with first new geometry here
from plotnine import geom_density

# Convert 'cyl' to string to treat it as a categorical variable
df['cyl'] = df['cyl'].astype(str)

plot_density = (
    ggplot(df, aes(x='mpg', fill='cyl')) +
    geom_density(alpha=0.6) +
    labs(title='Density Plot of mpg by cyl',
         x='mpg', y='Density')
)

print(plot_density)
```

```python
# Create plot with second new geometry here
from plotnine import geom_point

plot_scatter = (
    ggplot(df, aes(x='mpg', y='hp', color='cyl')) +
    geom_point() +
    labs(title='Scatterplot of mpg and hp by cyl',
         x='mpg', y='hp')
)

print(plot_scatter)
```

```python
# Create plot with third new geometry here
from plotnine import geom_line, aes, ggplot, labs

# Convert 'cyl' to string to treat it as a categorical variable
df['cyl'] = df['cyl'].astype(str)

# Create a line plot with 'mpg' on the x-axis, 'hp' on the y-axis, and 'cyl' as color aesthetic
plot_line = (
    ggplot(df.sort_values(by='mpg'), aes(x='mpg', y='hp', color='cyl', group='cyl')) +
    geom_line() +
    labs(title='Line Plot of hp by mpg colored by cyl',
         x='mpg', y='hp')
)

print(plot_line)
```
